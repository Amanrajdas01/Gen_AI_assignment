```python
from transformers import pipeline
a = pipeline(task='text-generation', model='distilgpt2')


task = "Generate a dataset related to data science topics, including data cleaning, feature engineering, model building, and evaluation metrics."


output = a(task, max_length=150)

print(output)
```

```python
GPT2 = pipeline(task='text-generation', model='gpt2')


b = "Generate detailed text content describing a data science trainer, their background, teaching approach, and the type of courses they offer. The text should be informative and engaging, suitable for a professional profile or a promotional material."


output = GPT2(b, max_length=150)

print(output)
```

```python
from transformers import pipeline
bart_generator = pipeline(task='text-generation', model="microsoft/phi-2", truncation=True)
task = "Summarize the key points on why Data Science is an exciting field for the future."
post = bart_generator(task, max_length=150)
print(post)
```
